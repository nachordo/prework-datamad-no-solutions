```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
import random

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
gestures = ['rock', 'paper', 'scissors']

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...

def define_rounds():
    #The function asks for an odd number
    rounds = int(input("Enter the number of rounds (odd): "))
    while rounds % 2 == 0:
        #In case of an invalid choice, continue asking until a valid input
        rounds = int(input("The number of wounds must be odd, type it again: "))
    
    return rounds


n_rounds = define_rounds()

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
def win_rounds(rounds):
    #This function computes the neccesary number of winned rounds to win the game
    rounds =  rounds // 2 + 1
    return rounds

rounds_to_win = win_rounds(n_rounds)

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.
def cpu_gesture(choices):
    return random.choice(choices)


# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.
def player_gesture(choices):
    #The function asks for the gesture among the possible ones
    player_gesture = input("Enter your choice among "+str(choices)+": ")
    
    while not(player_gesture in choices):
        #In case of an invalid input, continue asking
        player_gesture = input("Enter a valid choice: ")
    
    return player_gesture

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
def check_round(player_choice,cpu_choice):
    #To make a simple code, we assume that the CPU wins
    result_match = 1
    
    #If the choice is equal, we change the result to a tie
    if player_choice == cpu_choice:
        result_match = 0
    #If the choices makes the payer win the round, we modify the result
    elif (player_choice == 'rock') & (cpu_choice == 'scissors'):
        result_match = 2
    elif (player_choice == 'scissors') & (cpu_choice == 'paper'):
        result_match = 2
    elif (player_choice == 'paper') & (cpu_choice == 'rock'):
        result_match = 2
        
    return result_match
    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
def resolve_round(player_choice,cpu_choice,result_match,player_win,cpu_win):
    #Shows the choice of the player and the CPU
    print("Player: ",str(player_choice)," CPU: ",str(cpu_choice))
    #Checks whether it's a tie
    if result_match == 0:
        print("It's a tie")
    #Adds 1 to the CPU win counter if it wins
    elif result_match == 1:
        cpu_win +=1
        print("The CPU wins the round")   
    #Adds 1 to the player win counter otherwise
    else:
        player_win +=1
        print("The player wins the round")
    
    return player_win,cpu_win
    
# Create two variables that accumulate the wins of each participant
cpu_score = 0
player_score = 0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.
def core_game(choices,rounds,playerscore,cpuscore):
    rounds_to_win = win_rounds(rounds)
    counter = 1
    
    #While no one wins and while the maximum numer of rounds isn't reached 
    while (counter <= rounds) & (playerscore <= rounds_to_win) & (cpuscore <= rounds_to_win):
        #Prints the round number
        print("Round ",counter)
        #Ask the player to choose the gesture
        player_choice = player_gesture(choices)
        #Randomly chooses the CPU's gesture
        cpu_choice = cpu_gesture(choices)
        #Resolves the round
        result_match = check_round(player_choice,cpu_choice)
        #Updates the CPU/Player win count if neccesary
        playerscore,cpuscore = resolve_round(player_choice,cpu_choice,result_match,playerscore,cpuscore)
        #Goes to the next round
        counter += 1

    return playerscore,cpuscore


cpu_score, player_score = core_game(gestures,n_rounds,cpu_score, player_score)
    
# Print by console the winner of the game based on who has more accumulated wins
def print_results(playerscore,cpuscore):
    #The function simply prints the result of the game
    if playerscore == cpuscore:
        print("GAME TIED")
    if playerscore < cpuscore:
        print("WINNER: The CPU")
    if playerscore > cpuscore:
        print("WINNER: The Player")
    
    return

print_results(cpu_score, player_score)
```
